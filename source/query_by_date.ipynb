```python
import pandas as pd

# Load the Excel sheet into a DataFrame
df = pd.read_excel("11082025.xlsx", sheet_name="Detail", engine="openpyxl")
df.head(5)
```

```python
# Loop over the first column then when the cell starts with "SUM", get the index value
table_indices = []
for i, value in enumerate(df.iloc[:, 0]):
    if str(value).startswith("SUM"):
        table_indices.append(i)
table_indices
```

```python
# In df, loop over all values at the row index = 1 then match when value="Thực tế"
for col in df.columns:
    if df.at[1, col] == "Thực tế":
        print(f"Match found in column: {col}")
```

```python
# Build smaller dataframe restricted by indexes pair in table_indices
tables = []
for i in range(len(table_indices) - 1):
    start = table_indices[i]
    end = table_indices[i + 1]
    tables.append(df.iloc[start + 1:end])

# Convert the first column to datetime of format dd/mm/yyyy
for table in tables:
    table.iloc[:, 0] = pd.to_datetime(table.iloc[:, 0], format="%d/%m/%Y")
    # Change all the year to 2025, keep the date
    table.iloc[:, 0] = table.iloc[:, 0].apply(lambda x: x.replace(year=2025))
    table.reset_index(drop=True, inplace=True)

# Filter data by date
query_date = "11/08/2025"
# Scan over all tables to match the date
filtered_data = pd.concat([table[table.iloc[:, 0] == pd.to_datetime(query_date, format="%d/%m/%Y")] for table in tables])

# Extract value
gas_production = filtered_data.iat[0, filtered_data.columns.get_loc("Unnamed: 50")]
gas_sale = filtered_data.iat[0, filtered_data.columns.get_loc("Unnamed: 55")]
flare = filtered_data.iat[0, filtered_data.columns.get_loc("Unnamed: 59")]
oil = filtered_data.iat[0, filtered_data.columns.get_loc("Unnamed: 64")]

print(gas_production, gas_sale, flare, oil)
```
