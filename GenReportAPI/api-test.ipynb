```python
import requests

API_URL = "http://0.0.0.0:3339/report"   
POSTGRES_DB = "QLKTDB"
POSTGRES_USER = "dev"
POSTGRES_PASSWORD = "StR0&GP@sSW)R4"
HOST = 'host.docker.internal'
PORT = 5432

payload = {
    "query_date": "2025/09/02",
    "POSTGRES_DB": POSTGRES_DB,
    "POSTGRES_USER": POSTGRES_USER,
    "POSTGRES_PASSWORD": POSTGRES_PASSWORD,
    "HOST": HOST,
    "PORT": PORT
}
```

```python
import pandas as pd
# --- Get DataFrame as JSON ---
resp_df = requests.post(f"{API_URL}/oilreport", json=payload)
if resp_df.status_code == 200:
    data = resp_df.json()
    print("DataFrame JSON result:")
    print(data[:3])
else:
    print("Error:", resp_df.status_code, resp_df.text)
daily_oil_report = pd.DataFrame(data)
daily_oil_report
```
